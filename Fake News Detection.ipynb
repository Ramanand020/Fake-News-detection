```python
import numpy as np
import pandas as pd


import os

fake_data = pd.read_csv("archive (3)/Fake.csv")
true_data = pd.read_csv("archive (3)/True.csv")
```

```python
import re
```

```python
def wordopt(text):
    #lower case
    text = text.lower()
    #remove urls
    text = re.sub(r"http?://\S+|www\. \S+","",text)
    #remove html tags
    text = re.sub(r"<.*?>","",text)
    #remove puncts
    text = re.sub(r"[^\w\s]","",text)
    #remove digits
    text = re.sub(r"\d","",text)
    #remove newline chars
    text = re.sub(r"\n","",text)
    return text
```

```python
news["text"] = news["text"].apply(wordopt)
```

```python
x = news["text"]
y = news["label"]
```

```python
def output_label(n):
    if n==0:
        return "It is a Fake news"
    elif n==1:
        return "It is a Real news"
```

```python
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    pred_dtc = DTC.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    return "\n\nLR Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_lr[0]), output_label(pred_dtc[0]), output_label(pred_gbc[0]), output_label(pred_rfc[0]))
```
